```python
def dist(a1,b1,c1,d1,a2,b2,c2,d2):
    return np.sqrt((a1-a2)**2+(b1-b2)**2+(c1-c2)**2+(d1-d2)**2)
```

```python
def returncluster(m1,m2,m3,a,b,c,d):
    d1 = dist(m1[0],m1[1],m1[2],m1[3],a,b,c,d)
    d2 = dist(m2[0],m2[1],m2[2],m2[3],a,b,c,d)
    d3 = dist(m3[0],m3[1],m3[2],m3[3],a,b,c,d)

    if d1<d2 and d1<d3:
        return 1
    elif d2<d3 and d2<d1:
        return 2
    else:
        return 3
```
